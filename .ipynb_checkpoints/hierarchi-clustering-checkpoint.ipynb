```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import dataset
link = "https://raw.githubusercontent.com/iyanrspian/klasterisasi-hirarkis/main/wholesale-customers-data.csv"
data = pd.read_csv(link)
data.head()
```

```python
# Normalisasi data, agar semua variabel mempunyai skala yang sama
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
# Menampilkan dendrogram untuk mengetahui banyaknya cluster pada kasus ini
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
plt.ylabel('Jarak')
plt.xlabel('Pelanggan')
dend = sch.dendrogram(sch.linkage(data_scaled, method='ward'))

# X-axis berisi sampel, sedangkan Y-axis mewakili jarak antar sampel
# Garis vertikal merupakan pembatas dengan jarak maksimum pada garis biru
# Oleh karena itu, untuk memotong dendrogam kita tentukan dengan ambang batas '6'
plt.axhline(y=6, color='r', linestyle='--')
plt.show()
```

```python
# Terdapat 2 cluster, setelah melakukan proses 'cut' pada dendrogram
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_scaled)
```

```python
# Visualisasi
plt.scatter(data_scaled['Milk'], data_scaled['Grocery'], c=cluster.labels_)
plt.title('Cluster of Customers')
plt.xlabel('Milk')
plt.ylabel('Grocery')
plt.show()
```

```python
# Atau secara sederhana bisa seperti ini
X = data_scaled.iloc[:, [3,4]].values
y_hc = cluster.fit_predict(data_scaled)

plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 60, c = 'm', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 60, c = 'g', label = 'Cluster 2')
plt.title('Clusters of customers')
plt.xlabel('Milk')
plt.ylabel('Grocery')
plt.legend()
plt.show()
```
